# KYC Automation using OCR

## Objective
The objective of this project is to automate the KYC (Know Your Customer) process by extracting text from identity documents using OCR and validating key identity information such as PAN number. This helps reduce manual verification effort and improves efficiency in identity validation.

```python
import numpy as np
import cv2

img = np.ones((100, 400), dtype='uint8') * 255

cv2.putText(img, 'ABCDE1234F', (10, 60),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (0), 2, cv2.LINE_AA)

cv2.imwrite('pan_test.png', img)
```

```python
result = reader.readtext('pan_test.png')
text = " ".join([r[1] for r in result])
print(text)
```

```python
import re

pan = re.findall(r'[A-Z]{5}[0-9]{4}[A-Z]', text)

if pan:
    print("Valid PAN detected:", pan)
else:
    print("No valid PAN found")
```

```python
if pan:
    print("KYC Passed")
else:
    print("KYC Failed / Manual Review")
```
